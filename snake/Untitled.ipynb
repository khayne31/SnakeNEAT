```python
import networkx as nx
```

```python
import NEAT 
```

```python
network = NEAT.network()
```

```python
network.initalize_network(8, 4)
```

```python
print(network.input_nodes)
```

```python
G = nx.DiGraph()
```

```python
new_node = NEAT.node("test")
network.nodes[0].add_output_node(new_node, 2)
network.nodes.append(new_node)
```

```python
G.add_nodes_from(network.nodes)
```

```python
import matplotlib.pyplot as plt
```

```python
nx.draw(G, with_labels = True)
```

```python
G.clear()
for node in network.nodes:
    for connection in node.connected_to_out:
        G.add_edge(connection.input_node, connection.output_node, weight = connection.weight)
        #G.add_edge(connection.output_node, connection.input_node, weight = connection.weight)
        
labels = []
nodes = network.nodes
for node in nodes:
    labels.append(node.label)
mapping = dict(zip(nodes, labels))
G1 = nx.relabel_nodes(G, mapping)
```

```python
nx.draw_shell(G1, with_labels = True)
```
